```python
nul_val=[]
for i in list(data.columns):
    if data[i].isnull().sum() > 0:
        nul_val.append(i)
#         print(i, ': ', data[i].isnull().sum())
```

```python
categorical_val=[]
categorical_and_nul=[]

for i in list(data.columns):
    if data[i].dtype=="object" and data[i].isnull().sum() > 0:
        categorical_and_nul.append(i)
    if data[i].dtype=="object":
        categorical_val.append(i)   
print(len(categorical_and_nul)) 
categorical_and_nul
print(len(categorical_val))
categorical_val
        
```

```python
numerical_null=[]
for i in nul_val:
    if i not in  categorical_and_nul:
        numerical_null.append(i)
print(numerical_null)
```

```python
# handling numerical null values...................
data["LotFrontage"]=data["LotFrontage"].replace(np.nan,data["LotFrontage"].mean())
data["MasVnrArea"]=data["MasVnrArea"].replace(np.nan,data["MasVnrArea"].mean())
# data.drop(["GarageYrBlt"],axis=1)
# data.drop(["Id"],axis=1)
data.drop(columns=["GarageYrBlt","Id"],inplace=True)
```

```python
# handling which are categorical and nul values........................

for i in categorical_and_nul:
    arr=data[i].unique()
    data[i]=data[i].replace(np.nan,np.random.choice(arr))
```

```python
# handling categorical values

label_encoders={}

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in data.columns:
    if data[i].dtype=="object":
        label_encoders[i] = LabelEncoder()
        data[i] = label_encoders[i].fit_transform(data[i])
```
